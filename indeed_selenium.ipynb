```python
def data(y):
    x=y.find('h2',class_='jobsearch-JobInfoHeader-title css-161nklr e1tiznh50')
    title=x.get_text(strip=True).replace('- job post','')

    w=y.find('div',class_='css-1cjkto6 eu4oa1w0')
    company=w.get_text(strip=True)

    v=y.find('div',class_='css-6z8o9s eu4oa1w0')
    location=v.get_text(strip=True)

    job_url=driver.current_url



    match = re.search(r'&vjk=([^&]+)', job_url)
    job_code = match.group(1)

    external_url=f'in.indeed.com/applystart?jk={job_code}'

    text=y.get_text(strip=True)
    found_skills = [skill for skill in unique_skills_set if skill.lower() in text.lower()]



    description=y.prettify()

    dict={'Position':title,'company':company,'location':location,'job_url':job_url,'job_id':job_code,'external_apply':external_url,'skills':[found_skills],'description':description}
    return dict
```
